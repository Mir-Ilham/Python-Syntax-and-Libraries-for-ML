```python
import pandas as pd
india_weather = {
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
}

indf = pd.DataFrame(india_weather)

indf
```

```python
us_weather = {
    "city": ["new york", "chicago", "orlando"],
    "temperature": [23, 14, 35],
    "humidity": [68, 65, 75]
}

usdf = pd.DataFrame(us_weather)

usdf
```

```python
df = pd.concat([indf, usdf])
df
```

```python
df = pd.concat([indf, usdf], ignore_index=True)
df
```

```python
df = pd.concat([indf, usdf], keys=["india", "us"])
df
```

```python
df.loc["india"]
```

```python
temperature = {
    "city" : ["mumbai", "delhi", "banglore"],
    "temperature": [32, 45, 30],
}

tmpdf = pd.DataFrame(temperature, index=[0, 1, 2])

tmpdf
```

```python
windspeed = {
    "city" : ["mumbai", "delhi", "banglore"],
    "windspeed": [7, 12, 9],
}

windspeed2 = {
    "city" : ["delhi", "mumbai"],
    "windspeed": [7, 12],
}

wdf = pd.DataFrame(windspeed)

wdf2 = pd.DataFrame(windspeed2, index=[1, 0])

wdf
```

```python
df = pd.concat([tmpdf, wdf])
df
```

```python
df = pd.concat([tmpdf, wdf], axis=1)
df
```

```python
df = pd.concat([tmpdf, wdf2], axis=1)
df
```
