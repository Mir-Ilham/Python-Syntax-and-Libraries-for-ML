```python
from io import StringIO
```

```python
data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        'c,d,3')
```

```python
df = pd.read_csv(StringIO(data), usecols=['col1', 'col2', 'col3'], dtype={'col1': object, 'col2': object, 'col3': int})
```

```python
df['col1']
```

```python
df
```

## Read JSON files

```python
Data = StringIO('{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}')
pd.read_json(Data)
```

## Read HTML files

```python
url = "https://www.fdic.gov/bank/individual/failed/banklist.html"
dfs = pd.read_html(url)
```

```python
dfs[0]
```

```python
dfs[0].describe()
```
